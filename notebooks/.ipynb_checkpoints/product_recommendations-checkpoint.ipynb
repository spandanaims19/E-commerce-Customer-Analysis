```python
from src.data_processing import load_and_clean_data
from src.recommendations import create_purchase_matrix, generate_association_rules, get_recommendations
```

```python
print("Loading and cleaning the data......")
retail_data_clean=load_and_clean_data(r"D:\Coding\python\E-commerce_Customer_Segmentation\online+retail\Online Retail.xlsx")
```

```python
print("Creating purchase matrix......")
purchase_matrix=create_purchase_matrix(retail_data_clean)
print(f"Purchase Matrix Shape : {purchase_matrix.shape}")
purchase_matrix.head()
```

```python
purchase_matrix.info()
```

```python
print("Generating Association Rules......")
rules=generate_association_rules(purchase_matrix)
```

```python
if rules is not None:
    print(f"Number of rules generated : {len(rules)}")
    print("\nTop 10 rules by lift : ")
    print(rules.head(10))

    # Visualize top associations as a network graph
    print("\nVisualizing product associations......")

    # Create a network graph of associations
    G=nx.Graph()

    # Add edges for top 20 rules

    for _, row in rules.head(20).iterrows():
        for item1 in row["antecedents"]:
            for item2 in row["consequents"]:
                # Add nodes
                if item1 not in G:
                    product1=retail_data_clean[retail_data_clean["StockCode"]==item1]["Description"].iloc if item1 in retail_data_clean["StockCode"].values else f"Product {item1}"
                    G.add_node(item1, name=product1[:20])
                if item2 not in G:
                    product2=retail_data_clean[retail_data_clean["StockCode"]==item2]["Description"].iloc if item2 in retail_data_clean["StockCode"].values else f"Product {item2}"
                    G.add_node(item2, name=product2[:20])

                # Add edge with lift as weight
                G.add_edge(item1, item2, weight=row["lift"])

    plt.figure(figsize=(12,12))
    pos=nx.spring_layout(G, k=0.3)
    nx.draw(G, pos, with_labels=False, node_size=50, width=1, alpha=0.7)

    node_labels=nx.get_node_attributes(G, "name")
    nx.draw_networkx_labels(G, pos, labels=node_labels, font_size=8)

    plt.title("Product Association Network (Top 20 Rules)")
    plt.tight_layout()
    plt.show()

    print("\nExample Product Recommendations : ")
    customer_purchase_counts=purchase_matrix.sum(axis=1)
    sample_customers=[
    customer_purchase_counts.sort_values(ascending=False).index[0],
    customer_purchase_counts.sort_values().index[-20],
    customer_purchase_counts.sort_values().index[20]
    ]
    
    for customer_id in sample_customers:
        print(f"\nRecommendations for Customer {customer_id} : ")
        recommendations=get_recommendations(customer_id, purchase_matrix, rules, retail_data_clean)
        
        print(f"Items previously purchased : {purchase_matrix.loc[customer_id].sum()}")
        
        if(recommendations):
            for i, (stock_code, desc, lift) in enumerate(recommendations, 1):
                print(f"{i}. {desc[:50]} (Code: {stock_code}) - Lift : {lift:.2f}")
            
        else:
            print("No recommendations available.")

else:
    print("No association rules were generated. Try adjusting the support threshold.")
```
