```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.gridspec import GridSpec
```

```python
from src.data_processing import load_and_clean_data
from src.segmentation import calculate_rfm, segment_customers
from src.visualizations import plot_customer_segments
```

```python
sns.set_style("whitegrid")
sns.set_palette("viridis")
```

```python
print("Loading and cleaning the data......")
retail_data_clean=load_and_clean_data("online+retail/Online Retail.xlsx")
```

```python
print("Calculating RFM metrics......")
rfm=calculate_rfm(retail_data_clean)
```

```python
print("\nRFM Metrics Summary")
rfm.describe()
```

```python
fig, axes=plt.subplots(1, 3, figsize=(18, 6))
sns.histplot(rfm["Recency"], kde=True, ax=axes[0])
axes[0].set_title("Recency Distribution")
sns.histplot(rfm["Frequency"], kde=True, ax=axes[1])
axes[1].set_title("Frequency Distribution")
sns.histplot(rfm["Monetary"], kde=True, ax=axes[2])
axes[2].set_title("Monetary Distribution")
plt.tight_layout()
plt.show()
```

```python
print("\nPerforming customer segmentation")
rfm_segmented, cluster_avg=segment_customers(rfm)
```

```python
print("\nCluster Averages : ")
cluster_avg
```

```python
plot_customer_segments(rfm_segmented)
```

```python
print("\nDetailed Segment Profiles : ")
rfm_segmented["_count"]=1
segment_profiles=rfm_segmented.groupby("Segment").agg({"Recency":["mean", "min", "max"], "Frequency":["mean", "min", "max"], "Monetary":["mean", "min", "max"], "_count":"sum"})

# Rename the columns properly
# First convert to a list of tuples
column_tuples=list(segment_profiles.columns)
# Replace "_count" with "Count" in the first position of the tuple where required
new_columns=[("Count", "sum") if t[0]=="_count" else t for t in column_tuples]
segment_profiles.columns=pd.MultiIndex.from_tuples(new_columns)
segment_profiles
```
