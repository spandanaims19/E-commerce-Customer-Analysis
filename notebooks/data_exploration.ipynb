```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
from src.data_processing import load_and_clean_data
```

```python
sns.set_style("whitegrid")
sns.set_palette("viridis")
# print(plt.style.available) to know all the available matplot styles
```

```python
print("Loading the raw retail dataset......")
file_path="online+retail/Online Retail.xlsx"
retail_data_clean=load_and_clean_data(file_path)
retail_data_clean.head()
```

```python
print("\nDataset Overview\n")
print(f"Shape : {retail_data_clean.shape}")
print(f"Number of unique customers : {retail_data_clean["CustomerID"].unique()}")
print(f"Number of unique products : {retail_data_clean["StockCode"].unique()}")
print(f"Data Range : {retail_data_clean["InvoiceDate"].min()} to {retail_data_clean["InvoiceDate"].max()}")
```

```python
print("\nSample Data")
retail_data_clean.head()
```

```python
print("\nData Types")
retail_data_clean.dtypes
```

```python
print("\nMissing values count : ")
retail_data_clean.isnull().sum()
```

```python
print("\nBasic Statistics")
retail_data_clean.describe()
```

```python
plt.figure(figsize=(12,6))
retail_data_clean["InvoiceDate"].dt.date.value_counts().sort_index().plot()
plt.title("Number of Transactions over Time")
plt.xlabel("Date")
plt.ylabel("Number of Transactions")
plt.tight_layout()
plt.show()
```

```python
top_products=retail_data_clean.groupby("Description")["Quantity"].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(12, 6))
top_products.plot(kind="barh")
plt.title("Top 10 Products by Quantity Sold")
plt.xlabel("Quantity Sold")
plt.tight_layout()
plt.show()
```

```python
customer_spending=retail_data_clean.groupby("CustomerID")["TotalPrice"].sum()
plt.figure(figsize=(10, 6))
sns.histplot(customer_spending, kde=True)
plt.title("Distribution of Customer Spending")
plt.xlabel("Total Spending")
plt.tight_layout()
plt.show()
```
